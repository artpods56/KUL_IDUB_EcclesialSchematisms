```python
from shapely import wkt
import numpy as np

# Filter the CSV data for Włocławek 1872 entries
wloclawek_df = df[df['skany'] == 'wloclawek_1872'].copy()

if len(wloclawek_df) == 0:
    print("No entries found for wloclawek_1872 in the CSV file.")
    print("Available schematisms in the CSV:")
    print(df['skany'].unique())
else:
    # Convert the_geom strings to shapely geometries
    wloclawek_df['geometry'] = wloclawek_df['the_geom'].apply(wkt.loads)
    
    # Create a GeoDataFrame from the filtered CSV data
    gdf_entries = gpd.GeoDataFrame(wloclawek_df, geometry='geometry')
    
    # Now we have two GeoDataFrames:
    # 1. gdf - from the shapefile (contains page locations)
    # 2. gdf_entries - from the CSV (contains schematism entries)
    
    # Perform spatial join
    # This will match entries to pages based on spatial intersection
    joined = gpd.sjoin(gdf_entries, gdf, how="inner", predicate="intersects")
    
    # Display some statistics about the matching
    print(f"Total entries in Włocławek 1872: {len(wloclawek_df)}")
    print(f"Total pages in shapefile: {len(gdf)}")
    print(f"Total matches found: {len(joined)}")
    
    # Display the first few matches with relevant columns
    print("\nSample matches (first 5):")
    print(joined[['dekanat', 'parafia', 'wezwanie', 'material_typ', 'location']].head())
```
